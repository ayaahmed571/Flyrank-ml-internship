## 1. My lane as an ML task (type)

The primary ML task is **ranking/scoring**. The goal is to rank content pages by their likelihood of needing review, so editors can prioritize their limited time.

A classification model can estimate decline risk, but the final objective is to produce a ranked review queue rather than simply predicting a label.

## 2. Target or proxy

For the starter dataset, I will use **is_declining_label** as a teaching proxy target.

This label indicates whether a page is marked as declining in the starter dataset. It is useful for learning the workflow, although a stronger production target would be based on future observed outcomes rather than the current time window.

## 3. Success metric

The primary evaluation metric is **Precision@50**.

The goal is to maximize the proportion of truly relevant pages among the top 50 recommendations, since reviewers usually have limited capacity and cannot inspect every page.

## 4. The unit of analysis, as a real dataframe

One row represents one pseudonymized content page.

Each row contains observable search performance, freshness, engagement, and content characteristics collected for that page.

```python
import pandas as pd

df = pd.read_csv("data/raw/content_refresh_anonymized.csv")

print(df.shape)

df["is_declining_label"] = (df["trend_direction"] == "down").astype(int)

df[[
    "content_id",
    "impressions_90d",
    "avg_position",
    "ctr",
    "is_declining_label"
]].head()
```

## 5. Why ML beats a fixed rule here


A fixed rule can only combine a small number of manually defined conditions.

Machine learning can combine multiple observable signals such as impressions, CTR, average position, freshness, and engagement to prioritize pages more effectively.

The output supports human decision-making by helping reviewers decide which pages to inspect first.
